```python
from sklearn.tree import DecisionTreeClassifier #import libary scikit learn untuk tree
from sklearn.neighbors import KNeighborsClassifier #import libary scikit learn untuk k-NN
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.tree import export_graphviz
from six import StringIO
from IPython.display import Image 
from pydot import graph_from_dot_data
import pandas as pd
import numpy as np
```

```python
feature_names = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
data_training = [[5.1, 3.5, 1.4, 0.2], [4.9, 3, 1.4, 0.2],
                 [4.7, 3.2, 1.3, 0.2], [4.6,3.1,1.5,0.2],
                 [5,3.6,1.4,0.2], [7,3.2,4.7,1.4],
                 [6.4, 3.2, 4.5, 1.5], [6.9, 3.1, 4.9, 1.5],
                 [5.5, 2.3, 4, 1.3], [6.5, 2.8, 4.6, 1.5],
                 [6.3, 3.3, 6, 2.5], [5.8, 2.7, 5.1, 1.9],
                 [7.1,3,5.9,2.1], [6.3,2.9,5.6,1.8], [6.5,3,5.8,2.2]] #data training / training set 
label_training = ["Iris-setosa","Iris-setosa","Iris-setosa","Iris-setosa","Iris-setosa",
                  "Iris-versicolor","Iris-versicolor","Iris-versicolor","Iris-versicolor",
                  "Iris-versicolor","Iris-virginica","Iris-virginica","Iris-virginica",
                  "Iris-virginica","Iris-virginica"] #label training

tree = DecisionTreeClassifier() #pemanggilan fungsi decision tree
tree.fit(data_training, label_training) #proses training
knn = KNeighborsClassifier(n_neighbors=1) #pemanggilan fungsi k-NN dengan k=1
knn.fit(data_training, label_training) #proses training

#unlabelled = [[2,2,30]] #unlabelled data ppt slide 9
```

```python
dot_data = StringIO()
export_graphviz(tree, out_file=dot_data, feature_names=feature_names)
(graph, ) = graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```

```python
unlabelled = [[5.4, 3.9, 1.7, 0.4],[5.2, 2.7, 3.9, 1.4], [7.6, 3.0, 6.6, 2.1], [6.3, 2.7, 4.9, 1.8]]

hasilTree = tree.predict(unlabelled) #proses klasifikasi
print("-----------------------------------------")
print("hasil prediksi dari data diatas menggunakan algoritma tree adalah : ")
print("Unlabelled Data Result")
print("Data Pertama - Data Kedua - Data Ketiga - Data Keempat")
print(hasilTree)
print("dengan nilai akurasi nya adalah ", tree.score(data_training, label_training, sample_weight=None))
print("-----------------------------------------\n\n\n")
print("-----------------------------------------")
hasilKNN = knn.predict(unlabelled) #proses klasifikasi
print("hasil prediksi dari data diatas menggunakan algoritma K-NN adalah : ")
print("Unlabelled Data Result")
print("Data Pertama - Data Kedua - Data Ketiga - Data Keempat")
print(hasilKNN)
print("dengan nilai akurasi nya adalah ", knn.score(data_training, label_training, sample_weight=None))
print("-----------------------------------------")
```
